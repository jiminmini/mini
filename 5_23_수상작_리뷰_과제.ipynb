#**[주제 및 데이터]**
**주제**: 소셜 미디어 앱별 사용자 행동 분석: 어떤 앱 사용자가 더 많은 시간을 소비하고, 더 많이 포스팅하고, 좋아요 및 팔로우를 많이 하는가?


**데이터 설명**:
이 데이터는 1,000명의 사용자에 대해 다음 정보를 포함한다:


-User_ID: 사용자 고유 ID

-App: 사용 중인 소셜 미디어 앱 이름 (예: Facebook, Instagram, TikTok 등)

-Daily_Minutes_Spent: 하루 평균 사용 시간 (분)

-Posts_Per_Day: 하루 평균 포스트 수

-Likes_Per_Day: 하루 평균 좋아요 수

-Follows_Per_Day: 하루 평균 팔로우 수

---

#**[코드 리뷰]**
**1. EDA (탐색적 데이터 분석)**
1. 분포 확인:

- Daily_Minutes_Spent는 일부 사용자에서 매우 높게 나타나며, 전반적으로 정규분포보다는 오른쪽으로 긴 꼬리를 갖는 분포를 가짐.

- Likes_Per_Day, Follows_Per_Day 역시 유사한 분포 특성을 보임.

2. 앱별 사용 행태:

- 앱(App) 변수는 Facebook, Instagram, TikTok, Pinterest, LinkedIn 등으로 다양하며, 각 앱별로 활동량 차이가 존재.

3. 상관관계:

- Daily_Minutes_Spent와 Likes_Per_Day, Posts_Per_Day 간 양의 상관관계가 있음.

---
**2. 전처리**

1. 카테고리 인코딩:

- App 컬럼은 모델링을 위해 One-hot Encoding 혹은 Label Encoding 수행 필요.

2. 정규화/표준화:

- Daily_Minutes_Spent, Likes_Per_Day, Posts_Per_Day 등은 단위가 상이하므로 모델링 전에 표준화 작업 필요.

3. 이상치 처리:

- 사용 시간이 500분을 초과하는 경우는 이상치일 수 있어 Boxplot 및 IQR 등을 통해 제거하거나 로깅 처리.

---
**3. 모델링 (Modeling)**

1. 분류 문제 (예: 어떤 앱을 사용하는지 예측):

- 입력: 활동량 (Daily_Minutes_Spent, Posts_Per_Day, Likes_Per_Day, Follows_Per_Day)

- 출력: App

- 적용 가능한 모델: Random Forest, Decision Tree, Logistic Regression, SVM


2. 회귀 문제 (예: 사용 시간 예측):

- 입력: 활동량 (Posts_Per_Day, Likes_Per_Day, Follows_Per_Day)

- 출력: Daily_Minutes_Spent

- 적용 가능한 모델: Linear Regression, Ridge, Lasso, XGBoost


3. 모델 평가 지표:

- 회귀: RMSE, MAE

- 분류: Accuracy, F1-score, Confusion Matrix
