#**[주제 및 데이터]**
**주제**: 자율주행 차량의 센서 안테나 성능을 예측하는 문제로, 주어진 센서 데이터로부터 안테나의 성능 지표를 예측하는 회귀 분석 문제입니다.

**데이터 설명**:

1. Train 데이터: 센서의 다양한 특성과 해당 안테나의 성능 지표가 포함된 학습용 데이터셋입니다.



2. Test 데이터: 학습된 모델의 성능을 평가하기 위한 테스트용 데이터셋으로, 안테나 성능 지표는 포함되어 있지 않습니다.

각 데이터셋은 다양한 센서 특성(예: 주파수, 전력 등)을 포함하고 있으며, 이를 기반으로 안테나 성능을 예측해야 합니다.

#**[차별 점 및 배울 점]**

**✅ 차별점**
1. 시각화를 통한 인사이트 도출: EDA 단계에서 다양한 시각화 기법을 활용하여 데이터의 특성을 깊이 있게 분석하였습니다.

2. 다양한 전처리 기법 적용: 로그 변환, 다항 특성 생성 등 다양한 전처리 기법을 적용하여 모델의 성능을 향상시켰습니다.
