#**[주제 및 데이터]**
**주제**: Otto Group Product Classification Challenge는 주어진 제품 데이터를
      바탕으로 해당 제품이 어떤 카테고리에 속하는지 예측하는 문제입니다.

이 대회의 목적은 제품 분류 문제를 해결하는 머신러닝 모델을 구축하는 것입니다.

**데이터 설명**:

1. Train 데이터:

15,000개 이상의 제품에 대한 정보가 포함되어 있으며, 각 제품은 93개의 특성으로 설명됩니다.

목표 변수는 product_category (0~8)로, 총 9개의 카테고리 중 하나로 분류됩니다.

2. Test 데이터:

Test 데이터는 Train 데이터와 유사하지만 라벨이 없는 상태로 제공됩니다. 이 데이터에 대한 예측을 제출하여 리더보드 순위를 결정합니다.




---


#**[코드 리뷰]**
**1. EDA (탐색적 데이터 분석)**

**-기본 통계 분석:**

각 특성의 통계량을 확인하며 데이터의 분포, 결측값 등을 파악합니다.

pandas와 matplotlib를 사용해 각 특성의 분포를 시각화하고, 데이터의 특성을 이해합니다.



**-상관 행렬 시각화:**

특성 간의 상관 관계를 분석하여, 모델에 유용할 수 있는 특성 간의 관계를 파악합니다.


---


**2. 전처리**

**-결측값 처리:**

각 특성의 결측값을 처리하는 방법은 평균 대체, 중앙값 대체 등으로 이루어졌습니다.

이 방법을 통해 결측값을 최대한 최소화하며 모델의 성능을 유지할 수 있었습니다.

**-특성 변환:**

StandardScaler나 MinMaxScaler 등을 사용하여 특성들을 정규화합니다.

스케일링을 통해 모델이 잘 수렴하도록 도와줍니다.


---



**3. 모델링 (Modeling)**

XGBoost 모델을 사용하여, 예측 정확도를 높이는 데 주력했습니다.

XGBoost는 Gradient Boosting 알고리즘의 구현체로, 분류 문제에 매우 강력한 성능을 보입니다.

앙상블 기법을 사용하여 여러 개의 모델을 조합하여 성능을 개선했습니다.



<하이퍼파라미터 튜닝>

GridSearchCV나 RandomizedSearchCV를 활용하여 모델의 하이퍼파라미터를 최적화했습니다.

주요 하이퍼파라미터(예: learning_rate, max_depth 등)를 조정하여 성능을 향상시켰습니다.

교차 검증 (Cross-validation):

5-fold 교차 검증을 사용하여 모델의 성능을 안정적으로 평가하고, 오버피팅을 방지합니다.



---




#**[차별 점 및 배울 점]**


*   XGBoost 모델을 사용하여 높은 예측 성능을 얻은 점이 특징입니다. XGBoost는 Gradient Boosting 기법을 기반으로 하여 여러 모델을 앙상블하는 데 유리합니다.


*   5-fold 교차 검증을 통해 모델을 평가하고, 데이터의 분포에 관계없이 일반화 성능을 높일 수 있었습니다. 교차 검증 기법을 사용하여 모델의 **과적합(overfitting)**을 방지하고, 안정적인 예측을 할 수 있었습니다.

*   앙상블 기법의 중요성: 하나의 모델만 사용할 때보다 여러 모델을 결합했을 때 성능이 더 좋아진다는 점을 배웠습니다.

*   데이터 전처리와 EDA의 중요성: 데이터를 제대로 이해하고 전처리하는 것이 모델 성능에 큰 영향을 미친다는 것을 다시 한 번 깨달았습니다.
