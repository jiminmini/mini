#**[주제 및 데이터]**
**주제**: 은행 고객의 이탈 여부(Exited)를 예측하는 이진 분류 문제

**데이터 설명**:

- Client_ID: 고객 ID (개별 고객 식별용, 예측에 사용하지 않음)

- Exited: 이탈 여부 (0: 유지, 1: 이탈) → 예측 대상

- Account_Balance: 계좌 잔고

- Sex: 성별

- Full_Name: 고객 이름 (예측에 불필요)

- Product_Count: 사용 중인 제품 수






---


#**[코드 리뷰]**
**1. EDA (탐색적 데이터 분석)**

- .info()와 .describe()를 통해 기본적인 데이터 구조 및 통계량 파악

- 결측치가 존재하는지 확인

- Exited 값 분포를 countplot으로 시각화 → 불균형 데이터임을 확인

- 수치형 변수들의 상관관계를 heatmap으로 시각화

- Country별 이탈률과 Years_Old에 따른 이탈 분포도 시각적으로 확인함 (boxplot 등 사용)

- 나이 60세 이상이면서 이탈하지 않은 고객에 대한 조건 필터링 등 구체적 데이터 탐색 수행


---


**2. 전처리**
- 불필요한 변수(Client_ID, Full_Name) 제거

- LabelEncoder를 통해 범주형 변수 (Sex, Country) 인코딩

- StandardScaler로 수치형 변수 스케일링

- 타겟 불균형을 해결하기 위해 SMOTE 오버샘플링 적용
.


---



**3. 모델링 (Modeling)**

- 주요 모델: XGBoost, LightGBM, CatBoost

- 세 모델을 활용한 StackingClassifier 앙상블 구성

- StratifiedKFold(5-fold) 교차 검증 사용

- 각 모델별로 GridSearchCV를 사용해 하이퍼파라미터 튜닝 진행

- 최종적으로 스태킹 모델의 예측 결과를 submission.csv에 저장




---

✅ 차별 점

- SMOTE와 같은 오버샘플링 기법을 통해 데이터 불균형을 효과적으로 해결

- StratifiedKFold로 타겟 분포를 유지한 교차 검증 수행

- XGBoost, LGBM, CatBoost라는 강력한 모델들을 스태킹하여 예측 성능 극대화

- 단일 모델이 아닌 Stacking 기반 앙상블 전략 적용
